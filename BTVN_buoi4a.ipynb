```python
#Bài tập về nhà buổi 4: máy hút bụi với Model based reflex agent
import random

N = int(input("Nhập kích thước ma trận N: "))

floor = []

print("\nNhập ma trận sàn nhà:")
print("0 = sạch, 1 = bẩn")

for i in range(N):
    row = list(map(int, input(f"Dòng {i + 1}: ").split()))

    while len(row) != N or any(cell not in [0, 1] for cell in row):
        print("Dữ liệu không hợp lệ. Nhập lại!")
        row = list(map(int, input(f"Dòng {i + 1}: ").split()))

    floor.append(row)

x, y = 0, 0

#lưu trạng thái sàn nhà đã duyệt để tránh lặp lại
visited_states = set()

path_stack = []

def print_floor():
    print("\nTrạng thái sàn nhà:")
    for row in floor:
        print(row)

#kiểm tra sàn nhà
def is_clean():
    for row in floor:
        if 1 in row:
            return False
    return True

directions = [
    (-1, 0),  # lên
    (1, 0),   # xuống
    (0, -1),  # trái
    (0, 1)    # phải
]

step = 0

print("\n TRẠNG THÁI BAN ĐẦU ")
print_floor()

while not is_clean():

    print(f"\n BƯỚC {step} ")
    print(f"Vị trí hiện tại: ({x}, {y})")

    # Lưu trạng thái hiện tại
    visited_states.add((x, y))

    # Hút bụi nếu ô bẩn
    if floor[x][y] == 1:
        print("Ô hiện tại bẩn -> ĐANG HÚT BỤI")
        floor[x][y] = 0
    else:
        print("Ô hiện tại sạch")

    print_floor()

    possible_moves = []

    for dx, dy in directions:
        nx = x + dx
        ny = y + dy

        if 0 <= nx < N and 0 <= ny < N:

            if (nx, ny) not in visited_states:
                possible_moves.append((nx, ny))

    if possible_moves:

        path_stack.append((x, y))

        # Random trạng thái mới
        x, y = random.choice(possible_moves)

        print(f"Di chuyển đến: ({x}, {y})")

    else:
        # Không còn trạng thái mới
        print("Không còn trạng thái mới chưa duyệt!")

        if path_stack:
            x, y = path_stack.pop()
            print(f"Quay lui về: ({x}, {y})")
        else:
            print("Không thể di chuyển thêm!")
            break

    step += 1

print("\n HOÀN THÀNH ")
print("Sàn nhà đã được làm sạch!")
print(f"Tổng số bước: {step}")
```
