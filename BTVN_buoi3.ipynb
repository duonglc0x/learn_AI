```python

# -1 = vật cản
N = 3
room = []

print("Nhập ma trận 3x3:")
print("0 = bẩn")
print("1 = sạch")
print("-1 = vật cản")

for i in range(N):
    row = list(map(int, input(f"Hàng {i+1}: ").split()))
    room.append(row)

x, y = 0, 0

if room[x][y] == -1:
    print("Vị trí bắt đầu là vật cản!")
    exit()

visited = [[False]*N for _ in range(N)]

directions = [
    (-1, 0), 
    (1, 0),  
    (0, -1),  
    (0, 1)    
]

def print_room():
    print("\nTrạng thái phòng:")
    for r in room:
        print(r)
    print()


def clean(x, y):

    visited[x][y] = True

    print(f"\nRobot tại ({x},{y})")

   
    if room[x][y] == 0:
        print("Ô bẩn -> SUCK")
        room[x][y] = 1
    else:
        print("Ô sạch")

    print_room()

 
    for dx, dy in directions:

        nx = x + dx
        ny = y + dy

        if 0 <= nx < N and 0 <= ny < N:

            if room[nx][ny] != -1 and not visited[nx][ny]:

                print(f"Di chuyển tới ({nx},{ny})")

                clean(nx, ny)

print("\n BẮT ĐẦU DỌN DẸP ")

clean(x, y)

print("\n kết thúc dọn dẹp!")
print_room()
```
