```python
from collections import deque

class Node:
    def __init__(self, state, parent=None, action="", depth=0):
        self.state = state
        self.parent = parent
        self.action = action
        self.depth = depth

def matrix_to_string(state):
    return ''.join(str(num) for row in state for num in row)

def print_state(state):
    for row in state:
        print(*row)
    print()

def is_goal(state):
    goal = [
        [1, 2, 3],
        [4, 5, 6],
        [7, 8, 0]
    ]
    return state == goal

def find_zero(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j

def expand(node):

    children = []

    # L, R, U, D
    directions = [
        (0, -1, 'L'),
        (0, 1, 'R'),
        (-1, 0, 'U'),
        (1, 0, 'D')
    ]

    x, y = find_zero(node.state)

    for dx, dy, action in directions:

        nx = x + dx
        ny = y + dy

        if 0 <= nx < 3 and 0 <= ny < 3:

            new_state = [row[:] for row in node.state]

            # Hoán đổi
            new_state[x][y], new_state[nx][ny] = \
                new_state[nx][ny], new_state[x][y]

            child = Node(
                state=new_state,
                parent=node,
                action=action,
                depth=node.depth + 1
            )

            children.append(child)

    return children

def print_frontier(frontier):

    print("===== FRONTIER =====")

    if not frontier:
        print("Rong\n")
        return

    for i, node in enumerate(frontier):

        print(f"Node {i + 1}:")

        print("Action:", node.action)
        print("Depth:", node.depth)

        print_state(node.state)

def print_reached(reached):

    print("===== REACHED =====")

    if not reached:
        print("Rong\n")
        return

    for i, state_str in enumerate(reached):

        print(f"State {i + 1}: {state_str}")

    print()

def print_solution(goal_node):

    path = []

    current = goal_node

    while current:
        path.append(current)
        current = current.parent

    path.reverse()

    print("\n===== LOI GIAI =====")

    for i, node in enumerate(path):

        print(f"Buoc {i}")

        if node.action != "":
            print("Action:", node.action)

        print_state(node.state)

    print("Tong so buoc:", goal_node.depth)

def bfs(initial_state):

    frontier = deque()

    reached = set()

    root = Node(initial_state)

    frontier.append(root)

    reached.add(matrix_to_string(initial_state))

    step = 0

    while frontier:

        print("\n===================================")
        print("LAN DUYET:", step)
        print("===================================\n")

        current = frontier.popleft()

        print("===== NODE HIEN TAI =====")
        print("Action:", current.action)
        print("Depth:", current.depth)

        print_state(current.state)

        if is_goal(current.state):

            print(">>> DA TIM THAY LOI GIAI <<<")

            print_solution(current)
            return

        children = expand(current)

        for child in children:

            state_str = matrix_to_string(child.state)

            if state_str not in reached:

                frontier.append(child)

                reached.add(state_str)

        print_frontier(frontier)

        print_reached(reached)

        step += 1

    print("Khong tim thay loi giai!")

initial_state = []

print("Nhap ma tran 3x3 (0 la o trong):")

for i in range(3):
    row = list(map(int, input().split()))
    initial_state.append(row)

bfs(initial_state)
```
