```python
import random
import copy

print("Nhập trạng thái ban đầu của 8-Puzzle")
print("Ví dụ:")
print("1 2 3")
print("4 0 5")
print("6 7 8")

puzzle = []

for i in range(3):
    row = list(map(int, input(f"Dòng {i+1}: ").split()))

    while len(row) != 3:
        print("Nhập đúng 3 số!")
        row = list(map(int, input(f"Dòng {i+1}: ").split()))

    puzzle.append(row)

goal_state = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]

visited_states = set()

path_stack = []

def print_puzzle(state):
    print()
    for row in state:
        print(row)

def find_zero(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j

def state_to_tuple(state):
    return tuple(tuple(row) for row in state)

def is_goal(state):
    return state == goal_state

def get_possible_moves(state):

    moves = []

    x, y = find_zero(state)

    directions = [
        (-1, 0),  # lên
        (1, 0),   # xuống
        (0, -1),  # trái
        (0, 1)    # phải
    ]

    for dx, dy in directions:

        nx = x + dx
        ny = y + dy

        if 0 <= nx < 3 and 0 <= ny < 3:

            new_state = copy.deepcopy(state)

            new_state[x][y], new_state[nx][ny] = \
                new_state[nx][ny], new_state[x][y]

            moves.append(new_state)

    return moves

current_state = puzzle

step = 0

print("\n TRẠNG THÁI BAN ĐẦU")
print_puzzle(current_state)

while not is_goal(current_state):

    print(f"\n BƯỚC {step} ")

    print_puzzle(current_state)

    # lưu trạng thái hiện tại
    visited_states.add(state_to_tuple(current_state))

    # lấy các trạng thái có thể đi
    possible_states = get_possible_moves(current_state)

    # lọc trạng thái chưa đi
    new_states = []

    for state in possible_states:

        if state_to_tuple(state) not in visited_states:
            new_states.append(state)

    # nếu có trạng thái mới
    if new_states:

        # lưu để backtracking
        path_stack.append(copy.deepcopy(current_state))

        # chọn random
        current_state = random.choice(new_states)

    else:

        print("Không còn trạng thái mới!")

        # quay lui
        if path_stack:
            current_state = path_stack.pop()
            print("Đang backtracking...")
        else:
            print("Không tìm được lời giải!")
            break

    step += 1

if is_goal(current_state):

    print("\nĐÃ GIẢI THÀNH CÔNG")
    print_puzzle(current_state)
    print(f"Số bước: {step}")
```
