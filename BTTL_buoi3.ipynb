```python

N = int(input("Nhập kích thước ma trận (N x N): "))
room = []

print(f"Nhập ma trận {N}x{N} (0 = bẩn, 1 = sạch):")

for i in range(N):
    row = list(map(int, input(f"Hàng {i+1}: ").split()))
    room.append(row)

x = 0
y = 0

def print_room():
    print("\nTrạng thái phòng:")
    for r in room:
        print(r)
    print()

print("BẮT ĐẦU DỌN DẸP ")
while True:
    print_room()
    print(f"Robot đang ở vị trí ({x},{y})")
    if room[x][y] == 0:
        print("Ô bẩn -> THỰC HIỆN: SUCK")
        room[x][y] = 1
    else:
        print("Ô sạch")
        if x % 2 == 0:

            if y < N - 1:
                y += 1
                print("Di chuyển RIGHT")
            else:
                if x < N - 1:
                    x += 1
                    print("Di chuyển DOWN")
                else:
                    break
        else:
            if y > 0:
                y -= 1
                print("Di chuyển LEFT")
            else:
                if x < N - 1:
                    x += 1
                    print("Di chuyển DOWN")
                else:
                    break
print("\n HOÀN THÀNH ")
print_room()
```
